```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from datetime import datetime
```

```python
df = pd.read_csv('/content/Pais.csv')
df
```

```python
# Filtro 1: Los registros de 5 ‚Äúhost name‚Äù diferentes
unique_host_names = df['host_name'].unique()
selected_host_names = unique_host_names[:5]
filtro1 = df[df["host_name"].isin(selected_host_names)]
filtro1
```

```python
# Filtro 2: Los registros de los host que se hallan unido a Airbnb "antes del a√±o 2019"
df['host_since'] = pd.to_datetime(df['host_since'])
filtro2 = df[df["host_since"] < datetime(2019, 1, 1)]
filtro2
```

```python
# Filtro 3: Los registros de los host que responden "a m√°s tardar en 1 d√≠a"
filtro3 = df[df["host_response_time"].isin(["within an hour", "within a few hours", "within a day"])]
filtro3
```

```python
# Filtro 4: Los registros que tengan "al menos 5 accommodates"
filtro4 = df[df["accommodates"] >= 5]
filtro4
```

```python
# Filtro 5: Los registros de los tipos de cuarto: "Entire home" y "Private room"
filtro5 = df[df["room_type"].isin(["Entire home/apt", "Private room"])]
filtro5
```

```python
# Filtro 6: Los registros que "a lo mucho cuenten con 4 camas"
filtro6 = df[df["beds"].fillna(0) <= 4]
filtro6
```

```python
# Filtro 7: Los registros que "tengan 1 ba√±o o que sean superhost" pero con la condici√≥n de que "todos sean instant_bookable"
filtro7 = df[((df["bathrooms"].fillna(0) == 1) | (df["host_is_superhost"] == "t")) & (df["instant_bookable"] == "t")]
filtro7
```

```python
# Filtro 8: Los registros cuyo "review_scores_cleanliness sea menor de 3 y review_scores_rating sea mayor de 3"
filtro8 = df[(df["review_scores_cleanliness"].fillna(0) < 3) & (df["review_scores_rating"].fillna(0) > 3)]
filtro8
```

```python
filtro1.to_csv("Entregable1.csv")
filtro2.to_csv("Entregable2.csv")
filtro3.to_csv("Entregable3.csv")
filtro4.to_csv("Entregable4.csv")
filtro5.to_csv("Entregable5.csv")
filtro6.to_csv("Entregable6.csv")
filtro7.to_csv("Entregable7.csv")
filtro8.to_csv("Entregable8.csv")
```
