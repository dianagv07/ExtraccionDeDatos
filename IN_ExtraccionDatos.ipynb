```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_excel('/content/Detalle precios y productos fabricados 2022 (1).xlsx')
df
```

```python
df.info()
```

**Filtros para variables cualitativas**

```python
#Filtro por objeto
#Filtro por objeto, en este caso por el nombre del vendedor "DIEGO ARMANDO LEON DURAN"
filtro1=df[(df["NOMBRE_VENDEDOR"] == "DIEGO ARMANDO LEON DURAN")]
filtro1
```

```python
#Filtro por filas
#Filtro por filas, en este caso seleccionando las filas 2 y 5

#filtro2=df.iloc[3:11, :]
filtro2=df.iloc[[2,5],:]
filtro2
```

```python
#Filtro por columnas
#Filtro por columnas, en este caso seleccionando las columnas de la 1 a la 3
filtro3=df.iloc[:,1:3]
#filtro3=df.iloc[:,[1,4,5]]
filtro3
```

```python
df1 = pd.read_excel('/content/Detalle precios y productos fabricados 2022 (1).xlsx', index_col=3)
df1
```

```python
#Filtro por nombre de los indices, en este caso "SEVEN DAYS ALL SPORT" y "CPEG SOLUTIONS" y la columna "MARGEN_TOTAL_CALCULADO"
filtro4=df1.loc[["SEVEN DAYS ALL SPORT", "CPEG SOLUTIONS"], ["MARGEN_TOTAL_CALCULADO"]]
filtro4
```

```python
#filtro por cabecera
#filtro por cabecera, en este caso las primeras 5 filas
filtro5=df.head()
filtro5
```

```python
#filtro por cola
filtro6=df.tail()
filtro6
```

```python
#Filtro por comparacion
#Filtro por comparacion, en este caso los valores de la columna "COSTO_TOTAL_CALCULADO" mayores o iguales a 20000
filtro7=df[df["COSTO_TOTAL_CALCULADO"]>= 20000]
filtro7
```

```python
#Filtro Y
#Filtro Y, en este caso los valores de la columna "COSTO_TOTAL_CALCULADO" mayores a 15000 Y el nombre del vendedor igual a "ALFREDO CANELA"
filtro8=df[(df["COSTO_TOTAL_CALCULADO"]> 15000) & (df["NOMBRE_VENDEDOR"] == "ALFREDO CANELA")]
filtro8
```

```python
#Filtro O
#Filtro O, en este caso los valores de la columna "COSTO_TOTAL_CALCULADO" mayores a 5000 O el nombre del vendedor igual a "LETICIA RAMIREZ HERNANDEZ"
filtro9=df[(df["COSTO_TOTAL_CALCULADO"]> 5000) | (df["NOMBRE_VENDEDOR"] == "LETICIA RAMIREZ HERNANDEZ")]
filtro9
```

```python
#Filtro not
#Filtro not, en este caso los valores de la columna "COSTO_TOTAL_CALCULADO" que NO sean mayores a 5000 O el nombre del vendedor igual a "LETICIA RAMIREZ HERNANDEZ"
filtro10=df[~(df["COSTO_TOTAL_CALCULADO"]> 5000) | (df["NOMBRE_VENDEDOR"] == "LETICIA RAMIREZ HERNANDEZ")]
filtro10
```

```python
#filtro not
filtro10=df[~(df["CANT"] == "1")]
filtro10
```
