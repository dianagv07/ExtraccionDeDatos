```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_excel('/content/Datos de Facturación.xlsx')
df
```

```python
df.info()
```

```python
# Filtro 1: CVE_CLPV entre 1000 y 2000
df['CVE_CLPV'] = pd.to_numeric(df['CVE_CLPV'], errors='coerce')
df_filtrado_1 = df[(df['CVE_CLPV'] >= 1000) & (df['CVE_CLPV'] <= 2000)]
df_filtrado_1.to_csv("filtro_1_CVE_CLPV.csv", index=False)
print("Filtro 1 aplicado:", df_filtrado_1.head())
```

```python
# Filtro 2: CVE_VEND excepto "5" y "4"
df_filtrado_2 = df[~df['CVE_VEND'].astype(str).str.strip().isin(["5", "4"])]
df_filtrado_2.to_csv("filtro_2_CVE_VEND.csv", index=False)
print("Filtro 2 aplicado:", df_filtrado_2.head())
```

```python
# Filtro 3: FECHA_ENT igual a "28/02/2022"
fecha_filtro = pd.to_datetime("2022-02-28")
df_filtrado_3 = df[df['FECHA_ENT'] == fecha_filtro]
df_filtrado_3.to_csv("filtro_3_FECHA_ENT.csv", index=False)
print("Filtro 3 aplicado:", df_filtrado_3.head())
```

```python
# Filtro 4: CAN_TOT menor a 5951.7 o STATUS "E"
df_filtrado_4 = df[(df['CAN_TOT'] < 5951.7) | (df['STATUS'] == "E")]
df_filtrado_4.to_csv("filtro_4_CAN_TOT_STATUS.csv", index=False)
print("Filtro 4 aplicado:", df_filtrado_4.head())
```

```python
# Filtro 5: Solo columnas relevantes
df_filtrado_5 = df[['CVE_DOC', 'FECHA_ENT', 'FECHA_VEN', 'CAN_TOT']]
df_filtrado_5.to_csv("filtro_5_Columnas.csv", index=False)
print("Filtro 5 aplicado:", df_filtrado_5.head())
```

```python
# Filtro 6: Filas de 7001 a 7099
df_filtrado_6 = df.iloc[7001:7100]
df_filtrado_6.to_csv("filtro_6_Filas.csv", index=False)
print("Filtro 6 aplicado:", df_filtrado_6.head())
```

```python
# Filtro 7: Indexado por CVE_VEND (1,2) con FECHAELAB
df_filtrado_7 = df[df['CVE_VEND'].isin([1, 2])]
df_filtrado_7 = df_filtrado_7.set_index('CVE_VEND')[['FECHAELAB']]
df_filtrado_7.to_csv("filtro_7_Index.csv")
print("Filtro 7 aplicado:", df_filtrado_7.head())
```
